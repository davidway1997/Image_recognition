```python
# -*- coding: utf-8 -*-
"""
Created on Fri Sep 18 12:24:07 2020

@author: poww
"""

import numpy

data_points = numpy.array([[0.11, 1.24],  
                 [0.20, 1.40], 
                 [0.49, 2.00],
                 [0.65, 2.27],
                 [0.67, 2.41],
                 [0.83, 2.52]])

n_epoch = 4000
batch_size = 2
eta = 0.01

# Step 1: random initialization
a = numpy.random.randn(1)
b = numpy.random.randn(1)

# Step 2: repeatedly visit the entire data_points 
#           for n_epoch times (in this case=10)
for epoch in range(n_epoch):   
    # Step 2.1: shuffle the data points
    numpy.random.shuffle(data_points)
    
    # Step 2.2: for each mini-batch to update (a,b)
    for start_index in range(0, len(data_points), batch_size):
        points = data_points[start_index:start_index+batch_size]        
        x = points[:,0]
        y = points[:,1]
        
        # Step 2.2.1: compute the MSE loss L_MSE
        y_hat = a * x + b
        y_error = y - y_hat
        mse_loss = (y_error **2).mean()
        print(f'\n{a}:{b}:{mse_loss}')

        # Step 2.2.2: compute the gradients w.r.t. a and b
        a_grad = -2 * ((x * y_error).mean())
        b_grad = -2 * (y_error.mean())

        # Step 2.2.3: update model parameters a and b
        a = a - eta * a_grad
        b = b - eta * b_grad
```
